```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 1000)
```

# 데이터 불러오기

```python
# 데이터 불러오기
df_total = pd.read_csv('data/분양권/분양권_1619.csv')

# 중복데이터 제거
df_total.drop_duplicates(inplace=True)

# 불필요 컬럼 제거
df_total.drop(['id','거래유형','중개사소재지'],axis=1, inplace=True)
df_total
```

# 데이터 전처리

### 1. 열 이름 변경

```python
# 열 이름 변경
df_total.columns = ['시군구', '번지', '단지명', '전용면적_미제', '거래금액', '층', '계약년월', '계약일',
                '분/입구분', '해제사유발생일','지역']

# 전용면적 자료형을 실수로 변경
df_total['전용면적_미제'] = df_total['전용면적_미제'].astype(float)
df_total
```

### 2. 245$m^2$이상의 건물들 삭제

```python
# 245m2 이상의 건물들 삭제
over245_index = df_total[df_total['전용면적_미제']>=245].index
df_total.drop(over245_index,inplace =True)
df_total
```

### 3. 경기도 필터링

```python
# 경기도 필터링
def removeGyeonggi(df_total):
    df_total = df_total.copy()
    list_minus = ['광주시', '이천시', '안성시', '양평군', '여주시', '양주시', '포천시', '동두천시', '가평군', '연천군']

    for i in range(len(list_minus)):
        idx = df_total[df_total['시군구'].str.split(' ').str[1] == list_minus[i]].index
        df_total.drop(idx, inplace=True)
    return df_total

df_total = removeGyeonggi(df_total)
df_total
```

### 4. 해제사유발생일이 NaN이 아닌 데이터 삭제

```python
# 해제사유발생일이 NaN이 아닌 데이터 삭제 
def removeHaeje(df_total):
    df_total = df_total.copy()
    haeje = df_total[df_total['해제사유발생일'].isna()==False]    # 해제사유발생일

    df_total.drop(haeje.index.values, inplace=True)                 # 해제사유발생일 삭제
    df_total.reset_index(drop=True,inplace=True)                   # 인덱스 재정렬
    
    return df_total

df_total = removeHaeje(df_total)
df_total.drop('해제사유발생일', axis=1, inplace=True)
df_total
```

### 5. `new_Date`, `quarter` 변수 생성

```python
# new_Date 생성 함수
def newdate(df_total):
    df_total = df_total.copy()
    
    # Series 자료형을 Object로 변경
    df_total['계약년월'] = df_total['계약년월'].astype('O')
    df_total['계약일'] = df_total['계약일'].astype('O')
 
    l = []                                         # new_Date 데이터를 담을 리스트
    q = []                                         # quater 데이터를 담을 리스트
    for i in range(len(df_total.index.values)):
        year = str(df_total['계약년월'].iloc[i])[0:4]    # 년
        month = str(df_total['계약년월'].iloc[i])[4:]    # 월
        day = str(df_total['계약일'].iloc[i])            # 일
        
        # new_Date 리스트에 'yyyy-MM-dd' 형태로 데이터 추가
        l.append(year+'-'+month+'-'+day)
        
        # month를 숫자로 바꿔서 부등호조건 활용해서 쿼터 나누고 리스트에 담기.
        month = int(month)
        if month < 4:
            q.append(year+'_1Q')
        elif month < 7:
            q.append(year+'_2Q')
        elif month < 10:
            q.append(year+'_3Q')
        elif month < 13:
            q.append(year+'_4Q')
            
    # 파생변수 new_Date 생성        
    df_total['new_Date'] = l
    df_total['new_Date'] = pd.to_datetime(df_total['new_Date'])    # timestamp 자료형 전환
    
    # 파생변수 quater 생성
    df_total['quarter'] = q
    
    return df_total

df_total = newdate(df_total)
df_total
```

### 6. `자치구_동` 컬럼 생성

```python
# 자치궈_동 생성
def makeDong(df_total):
    df_total = df_total.copy()
    df_total['자치구_동'] = 'NaN'

    df_total['자치구_동'] = df_total['시군구'].str.split(' ').str[3]
    df_total.loc[df_total['자치구_동'].isna() == True,'자치구_동'] = df_total['시군구'].str.split(' ').str[2]

    return df_total

# 매매, 전월세 데이터에 자치구_동 생성
df_total = makeDong(df_total)
```

### 7. `전용면적_미제` 올림처리

```python
# 전용면적_미제 올림처리
from math import ceil

def ceilArea(df):
    df = df.copy()
    df['전용면적_미제'] = df['전용면적_미제'].apply(lambda x: ceil(x))
    df['전용면적_미제']

    return df

df_total = ceilArea(df_total)
df_total
```

### 8. `id` = `시군구` + `단지명` + `전용면적_미제`

```python
# 시군구, 번지, 단지명, 전용면적_미제로 id 변수 파생
def makeID(df):
    df = df.copy()
    df['id'] = df.시군구 +'_'+ df.단지명 +'_'+ df.전용면적_미제.astype(str)
    return df

df_total = makeID(df_total)
df_total
```

### 9. 거래금액 백만단위로 설정

```python
# 거래금액 백만단위로 조정
def changeMillion(df_total, keyword='매매'):
    df_total = df_total.copy()
    if keyword == '매매':
        df_total['거래금액'] = df_total['거래금액'].str.replace(',','')
        df_total['거래금액'] = df_total['거래금액'].astype(int)
        df_total['거래금액'] = df_total['거래금액']/100
    return df_total
df_total = changeMillion(df_total)

df_total
```

### 10. `분/입구분` 삭제

```python
# 분/입구분 삭제
df_total.drop('분/입구분',axis=1, inplace=True)
df_total
```

### 11. `지번주소` 컬럼 생성

```python
def bonbuToBeonji(df):
    df = df.copy()
    df['지번주소'] = 'NaN'
    
    df.지번주소 = df.시군구 + ' ' + df.번지
    
    return df

df_total= bonbuToBeonji(df_total)
df_total
```

### 12. `전용면적_미제`로부터 `전용면적_평` 변수 파생.

```python
# 미터제곱 전용면적을 평수로 바꿔 변수로 저장
def meterToPyeong(df):
    df = df.copy()
    df['전용면적_평'] = df['전용면적_미제'] / 3.305785
    df['전용면적_평'] = df['전용면적_평'].round()
    return df

df_total = meterToPyeong(df_total)
df_total
```

```python
# 열 재정렬
df_total = df_total[['시군구', '번지', '단지명', '전용면적_미제', '계약년월', '계약일', '거래금액', '층', '지역', '전용면적_평', '지번주소', 'new_Date', 'quarter', '자치구_동', 'id']]
```

```python
# 데이터 저장
df_total.to_csv('data/분양권/비인도데프_분양데이터.csv', index=False, encoding='utf-8-sig')
```

# 분양권 데이터 Grouping

```python
# id, quarter 별로 grouping
df_g = df_total.groupby(['id','quarter'],as_index=False).mean()
df_g
```

```python
from tqdm import tqdm
df_temp = df_g.drop_duplicates(['id'],keep='first')

df_list = []

# 모든 id에 각각 2017_1Q~2021_2Q를 넣은 테이블을 생성하는 함수
def make_quater(df):
    quater_list = ['2016_1Q', '2016_2Q', '2016_3Q', '2016_4Q',
                   '2017_1Q', '2017_2Q', '2017_3Q', '2017_4Q',
                   '2018_1Q', '2018_2Q', '2018_3Q', '2018_4Q',
                   '2019_1Q', '2019_2Q', '2019_3Q', '2019_4Q']
    for item in tqdm(quater_list):
        globals()['df_{}'.format(item)] = df.copy()
        globals()['df_{}'.format(item)]['quarter'] = str(item)
        df_list.append(globals()['df_{}'.format(item)])
        
    return

# df_temp에 함수 적용
make_quater(df_temp)

# df_2017_1Q ~ df_2019_4Q 테이블 이어붙인 df_temp2 테이블 생성
df_temp2 = pd.concat(df_list,axis=0)
# df_temp2[['보증금','거래금액','주택담보대출금리','월세','전월세전환율','전세보증금']] = 0

# 여기서 id, quater만 뽑으면 id에 맞게 2017_1Q ~ 2021_2Q가 들어있는 데이터 완성.
df_temp2 = df_temp2[['id','quarter']]
df_temp2
```

```python
# df_temp2 테이블에 df_mae_g를 right join
df_merge = pd.merge(df_g,df_temp2[['id','quarter']], how='right',on=['id','quarter'])
display(df_merge)
# id, quater 기준으로 재배열
df_merge = df_merge.sort_values(['id','quarter'])
df_merge.reset_index(drop=True, inplace=True)
display(df_merge)
# df_merge_2 = df_merge_2.reset_index()
```

```python
# 데이터 저장
df_merge.to_csv('data/분양권/비인도데프_분양데이터.csv', index=False, encoding='utf-8-sig')
```
